### 3. Preparing the data

```python
# Counting null values
ds.isnull().sum()
```

```python
ds.nunique().sort_values(ascending = False).values
```

```python
# Visualizing the above information in graphical mode
fig, ax = plt.subplots()

ax.bar(ds.nunique().sort_values(ascending = False).index, ds.nunique().sort_values(ascending = False).values, color = '#ADD8E6')

ax.tick_params(axis = 'x', labelrotation = 90)

plt.show()
```

```python
# Verifying the life_expect histogram
fig, ax = plt.subplots()

ax.hist(ds.Life_expect, bins=20, color = '#ADD8E6')

plt.show()
```

### Boxplot Visualization
